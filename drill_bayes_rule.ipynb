### Baye's Rule:

The probability of A given B equals the probability of B given A, times the probability of A, divided by the probability of B.

P(A|B) = P(B|A) \* P(A) / P(B)

We expand the probability of B in the second formula where A~ is the inverse of A, so in our case not being infected.

P(A|B) = P(B|A) \* P(A) / [P(A) \* P(B|A) + P(A~) \* P(B|A~)]

P(Infected | Positive Test) = P(Positive Test | Infected) \* P(Infected) / P(Positive Test)
