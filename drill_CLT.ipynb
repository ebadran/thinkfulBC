Create two binomially distributed populations with n equal to 10 and size equal to 10000. The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5. Using a sample size of 100, calculate the means and standard deviations of your samples.

```python
import numpy as np
import pandas as pd
import scipy
# from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

# Letâ€™s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()

print("pop1 mean:", pop1.mean())
print("pop2 mean:", pop2.mean())
print("pop1 standard deviation:", pop1.std())
print("pop2 standard deviation:", pop2.std())

# Compute the difference between the two population means.
diff=pop2.mean( ) - pop1.mean()
print('difference:', diff)
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, color='olive', label='sample 1') 
plt.hist(sample2, alpha=0.5, color='gold', label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print("sample_1 mean:", sample1.mean())
print("sample_2 mean:", sample2.mean())
print("sample_1 standard deviation:", sample1.std())
print("sample_2 standard deviation:", sample2.std())

# Compute the difference between the two sample means.
diff=sample2.mean( ) - sample1.mean()
print('difference:', diff)
```
