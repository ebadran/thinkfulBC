### Loading the data, fixing column names, and filling NANs:

```python
data = pd.read_excel('fbi_ny_2013.xls', sheet_name=0, header=1, skiprows=3, skip_footer=3,
                     index_col=0, thousands=None)

data.columns = ['population', 'violent_crime',
       'murder',
       'rape', 'rape_legacy',
       'robbery', 'assault', 'property_crime', 'burglary',
       'larceny_theft', 'vehicle_theft', 'arson']
data.index.names = ['city']
data = data.fillna(0)
data.head(n=2)
```

### Checking for anomalous cases:

```python
print(data.murder.dtype, data.robbery.dtype, data.population.dtype)
```

```python
data[['murder','robbery','population']].describe()
```

```python
p_crime = data[['population','murder','robbery']]
binarizer = Binarizer(threshold=0.0, copy=False).fit(p_crime[['murder','robbery']])
p_crime[['murder','robbery']] = binarizer.transform(p_crime[['murder','robbery']])
p_crime['population^2'] = p_crime.population**2
```
