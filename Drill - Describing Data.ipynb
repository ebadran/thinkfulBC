# Drill - Describing Data

```python
import numpy as np
import pandas as pd
import statistics
```

### Q1.
Greg was 14, Marcia was 12, Peter was 11, Jan was 10, Bobby was 8, and Cindy was 6 when they started playing the Brady kids on The Brady Bunch. Cousin Oliver was 8 years old when he joined the show. What are the mean, median, and mode of the kids' ages when they first appeared on the show? What are the variance, standard deviation, and standard error?

```python
df = pd.DataFrame()
df['Name'] = 'Greg', 'Marcia', 'Peter', 'Jan', 'Bobby', 'Cindy','Oliver'
df['Age'] = 14, 12, 11, 10, 8, 6, 8
df
```

```python
print("Mean = ", df.Age.mean())
print("Median = ", df.Age.median())
print("Mode = ", statistics.mode(df['Age']))
print("Variance = ", df.Age.var())
print("Standard deviation =", df.Age.std())
print("Standard Error = ", df.Age.sem())
```

### Q2.
Using these estimates, if you had to choose only one estimate of central tendency and one estimate of variance to describe the data, which would you pick and why?

Answers:
- For central tendency, I would choose the mean because there are no outliers in the dataset.
- For variance, I would choose the standard deviation, as it shows how far the values are from the mean. The standard deviation is expressed in the same unit as the values, which makes it easier to understand.

### Q3.
Next, Cindy has a birthday. Update your estimates - what changed, and what didn't?

Answer:
- The mode and median remained the same. All other measures changed, since they are the result of calculations.

```python
df.set_value(5, 'Age', 7)
print("Mean = ", df.Age.mean())
print("Median = ", df.Age.median())
print("Mode = ", statistics.mode(df['Age']))
print("Variance = ", df.Age.var())
print("Standard deviation =", df.Age.std())
print("Standard Error = ", df.Age.sem())
```

### Q4. 
Nobody likes Cousin Oliver. Maybe the network should have used an even younger actor. Replace Cousin Oliver with 1-year-old Jessica, then recalculate again. Does this change your choice of central tendency or variance estimation methods?

Answer:
- I would continue to use the mean and standard deviation. Despite Jessica's age being an outlier, the mean continues to be very similar to the median.

```python
df.loc[lambda df: df['Name']=='Oliver', :] = 'Jessica',1
print("Mean = ", df.Age.mean())
print("Median = ", df.Age.median())
# print("Mode = ", statistics.mode(df['Age']))
print("Variance = ", df.Age.var())
print("Standard deviation =", df.Age.std())
print("Standard Error = ", df.Age.sem())
```

### Q5.
On the 50th anniversary of The Brady Bunch, four different magazines asked their readers whether they were fans of the show. The answers were: TV Guide 20% fans Entertainment Weekly 23% fans Pop Culture Today 17% fans SciPhi Phanatic 5% fans. Based on these numbers, what percentage of adult Americans would you estimate were Brady Bunch fans on the 50th anniversary of the show?

Answer:
- I would estimate that the number of adult American fans would be between 12 and 20 fans.

```python
df = pd.DataFrame()
df['Magazine'] = 'TV Guide', 'Entertainment Weekly', 'Pop Culture Today', 'SciPhi Phanatic'
df['Fans'] = 20, 23, 17, 5

print("Mean = ", df.Fans.mean())
print("Median = ", df.Fans.median())
# print("Mode = ", statistics.mode(df['Fans']))
print("Variance = ", df.Fans.var())
print("Standard deviation =", df.Fans.std())
print("Standard Error = ", df.Fans.sem())

df
```
