```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

The **Pareto distribution** must be greater than zero, and is unbounded above. It is also known as the “80-20 rule”. In this distribution, 80 percent of the weights are in the lowest 20 percent of the range, while the other 20 percent fill the remaining 80 percent of the range.

```python
pareto = np.random.pareto(1000,100)

plt.hist(pareto)

# Add a vertical line at the mean.
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(pareto.mean() + pareto.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(pareto.mean() - pareto.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

The **Zipf distribution** (also known as the zeta distribution) is a continuous probability distribution that satisfies Zipf’s law: the frequency of an item is inversely proportional to its rank in a frequency table.

It is named for the American linguist George Kingsley Zipf, who noted that the frequency of any word in a sample of a language is inversely proportional to its rank in the frequency table.

```python
zipf = np.random.zipf(2.,100)

plt.hist(zipf)

# Add a vertical line at the mean.
plt.axvline(zipf.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(zipf.mean() + zipf.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(zipf.mean() - zipf.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

The **exponential distribution** is a continuous analogue of the geometric distribution. It describes many common situations, such as the size of raindrops measured over many rainstorms, or the time between page requests to Wikipedia.

```python
exponential = np.random.exponential(100,100)

plt.hist(exponential)

# Add a vertical line at the mean.
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(exponential.mean() - exponential.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

The **Logistic distribution** is used in Extreme Value problems where it can act as a mixture of Gumbel distributions, in Epidemiology, and by the World Chess Federation (FIDE) where it is used in the Elo ranking system, assuming the performance of each player is a logistically distributed random variable.

```python
logistic = np.random.logistic(10,1, 100)

plt.hist(logistic)

# Add a vertical line at the mean.
plt.axvline(logistic.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(logistic.mean() + logistic.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(logistic.mean() - logistic.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

The **multivariate normal, multinormal or Gaussian distribution** is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or “center”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution.

```python
mean = [0, 1]
cov = [[1, 0], [0, 10]]

multivariate_normal = np.random.multivariate_normal(mean,cov,100)

plt.hist(multivariate_normal)

# Add a vertical line at the mean.
plt.axvline(multivariate_normal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(multivariate_normal.mean() + multivariate_normal.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(multivariate_normal.mean() - multivariate_normal.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

The **Gumbel distribution** (named for German mathematician Emil Julius Gumbel) was used very early in the hydrology literature, for modeling the occurrence of flood events. It is also used for modeling maximum wind speed and rainfall rates. It is a “fat-tailed” distribution - the probability of an event in the tail of the distribution is larger than if one used a Gaussian, hence the surprisingly frequent occurrence of 100-year floods. Floods were initially modeled as a Gaussian process, which underestimated the frequency of extreme events.

```python
gumbel = np.random.gumbel(0.1, 10, 100)

plt.hist(gumbel)

# Add a vertical line at the mean.
plt.axvline(gumbel.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(gumbel.mean() + gumbel.std(), color='b', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(gumbel.mean() - gumbel.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

### Additionally:

1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
2. Add them together to create a third variable.
3. Graph the third variable using a histogram.
4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.
5. Evaluate the descriptive statistics against the data.


```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The values on the third quartile of the var1 distribution will overlap with the values on the first quartile of the var2 distribution. Therefore, adding them together generates a skewed distributution.
